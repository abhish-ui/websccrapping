```python
url=requests.get("https://stackoverflow.com/questions/415511/how-do-i-get-the-current-time-in-python",headers={"user-Agent":"mozilla/5.0"})
```

```python
soup=BeautifulSoup(url.text,"html.parser")
print(soup)
```

```python
question=soup.find("a",class_="question-hyperlink")
print(question.text)
```

```python
answer=soup.find_all("div",class_="s-prose js-post-body")
for i in answer:
  print(i.text)
```
